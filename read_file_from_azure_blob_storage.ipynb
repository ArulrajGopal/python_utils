```python
from azure.storage.blob import generate_blob_sas, BlobSasPermissions
from datetime import datetime, timedelta
import pandas as pd
```

```python
#enter credentials
account_name = 'arulrajstorageaccount'
account_key = ""
```

```python
def read_file(container_name, path):
    sas_i = generate_blob_sas(account_name = account_name,
                                container_name = container_name,
                                blob_name = path,
                                account_key=account_key,
                                permission=BlobSasPermissions(read=True),
                                expiry=datetime.utcnow() + timedelta(hours=1))

    sas_url = 'https://' + account_name+'.blob.core.windows.net/' + container_name + '/' + path + '?' + sas_i

    df = pd.read_csv(sas_url)

    return df
```

```python
df = read_file("public", "companies/companies_market_cap.txt")
```
